```python
import pandas as pd 
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns= {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create a column age_groupabs
labels = ['Young Adult', 'Adult','Middle-Age','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df.head(5)
```

```python
# Create column purchase_frequency_days 
frequency_mapping = {
    'Weekly':7, 'Fortnightly':14,'Bi-Weekly':14, 'Monthly':30, 'Quarterly':90,'Every 3 Month': 90,'Annually':365
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df.head()
```

```python
df['purchase_frequency_days'] = df['purchase_frequency_days'].fillna(0).astype(int)
```

```python
df.head()
```

```python
# Let us check of promo_code_used and discount_applied has the same value or different, so we can make to leave or delate one
df[['discount_applied','promo_code_used']].head()
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# as they contain the same value, we better delate one of them
df = df.drop(['promo_code_used'],axis = 1)
```

```python
df.to_csv('customer_shopping_behavior_clean.csv', index=False)
```
